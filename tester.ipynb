```python
data = pd.read_feather("data/spx.feather").reset_index()[["Date", "Adj Close"]]
data["rate"] = data["Adj Close"] / data["Adj Close"].shift(1)
data["rate"] = data["rate"].fillna(1.0)

hist = np.histogram(data["rate"], bins=1000, density=True)
dist = stats.rv_histogram(hist, density=True)

s = dist.stats(moments="mvsk")

pdf = extras.pdf_moments(s)

x = np.linspace(0.9, 1.1, 100)
fig = make_subplots()
fig.add_trace(
    go.Scatter(x=x, y=pdf(x))
)
fig.add_trace(
    go.Scatter(x=x, y=dist.pdf(x))
)
```

```python
def integrate(x, y):
    area = 0.0
    for i in range(len(x)-1):
        area += (y[i+1] + y[i]) / 2.0 * (x[i+1] - x[i])
    return area

def test_pdf(x, m, v, a):
    xx = x * x
    results = 2.0 * stats.norm.pdf(xx) * stats.norm.cdf(a*xx)
    return results / integrate(x, results)

x = np.linspace(-10.0, 10.0, 100)
fig = make_subplots()

for a in np.linspace(0.0, 10.0, 5):
    results = test_pdf(x, 1, 1, a)
    fig.add_trace(
        go.Scatter(x=x, y=results)
    )

    print(integrate(x, results))

fig.show()
```
